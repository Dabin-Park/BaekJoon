```python
def sol():
    for _ in range(int(input())):
        T = int(input())
        Woods = sorted(list(map(int,input().split())))
        New_Woods = [0]*T
        j = 0
        for i in range(T):
            New_Woods[(-1)**(i%2)*j - i%2] = Woods[i]
            j += i%2
        print(max([abs(New_Woods[i]-New_Woods[i-1]) for i in range(T)]))
            
sol()
```

```python
def sol():
    for _ in range(int(input())):
        T = int(input())
        Woods = sorted(list(map(int,input().split())))
        print(max([Woods[i+2] -Woods[i] for i in range(T-2)]))
            
sol()
```
